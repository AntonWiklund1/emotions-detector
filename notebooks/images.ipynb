```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../all_frames.csv', header=None)
```

```python

# Select which frame to display (e.g., the first frame)
frame_data = df.iloc[0]

# Convert the flat array back to a 2D array (48x48)
image = np.array(frame_data).reshape(48, 48)

# Display the image using matplotlib
plt.imshow(image, cmap='gray')
plt.title("Frame from CSV")
plt.axis('off')  # Turn off axis numbers and ticks
plt.show()
```
