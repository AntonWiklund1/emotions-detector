```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../all_frames.csv', header=None)

for index, row in df.iterrows():
    frame_data = row.values
    image = np.array(frame_data).reshape(48, 48)
    plt.imshow(image, cmap='gray')
    plt.title(f"Frame {index+1}")
    plt.axis('off')
    plt.show()
```
